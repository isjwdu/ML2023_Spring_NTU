```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_train = pd.read_csv('covid_train.csv')
useful_data = data_train.drop(data_train.columns[0:35],axis=1)

# 相關係數圖
def draw_corr_picture(X):   
    corrmat = X.corr()
    plt.subplots(figsize=(15,15))
    sns.heatmap(corrmat,vmax=0.9,square=True,cmap='Blues')
    plt.show()

draw_corr_picture(useful_data)
# By comparing the labels of each column with the last row, features with darker labels are considered highly correlated and selected as training features
# feature = [70,52,47,48,65,37,66,55,83,84,73,36,35,53,54,71,72]
```
